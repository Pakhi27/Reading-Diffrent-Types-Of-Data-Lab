```python
import numpy as np
from PIL import Image

def image_to_vector(image_path):
  img = Image.open(image_path)
  img_array = np.array(img)
  vector = img_array.flatten()
  return vector, img_array  # Return both vector and img_array

image_path = "/content/DS_1.jpeg"
vector, img_array = image_to_vector(image_path)  # Unpack the returned values
print(vector)

plt.plot(vector)
plt.show()
plt.imshow(img_array)
plt.show()  # Call plt.show() twice to display both plots
```

```python
# Color Histograms:
# Calculate the distribution of colors in the image.
# hist = cv2.calcHist([img], [0], None, [256], [0, 256]): Calculates the histogram of the image.
# [img]: The input image as a list.
# [0]: Specifies that we are calculating the histogram for the first channel (usually the blue channel).
# None: Specifies no mask, meaning the entire image is considered.
# [256]: Specifies the number of histogram bins (256 for 8-bit images).
# [0, 256]: Specifies the range of pixel values (0 to 255).
import numpy as np
import cv2

def image_to_histogram(image_path):
    img = cv2.imread(image_path)
    hist = cv2.calcHist([img], [0], None, [256], [0, 256])
    return hist.flatten()

image_path="/content/DS_1.jpeg"
histogram=image_to_histogram(image_path)
print(histogram)

plt.plot(histogram)
plt.show()
# X-axis: Represents the intensity values (bins) of the pixels in the image. For a grayscale image, it ranges from 0 (black) to 255 (white), with shades of gray in between.
# Y-axis: Represents the frequency or count. For each intensity value on the x-axis, the y-axis shows how many pixels in the image have that intensity.
```

```python
# Feature Extraction with Libraries:
# img = cv2.imread(image_path, 0): Reads the image from the specified image_path as grayscale. The 0 flag indicates grayscale loading.
# hog = cv2.HOGDescriptor(): Creates a HOGDescriptor object, which is used to compute the Histogram of Oriented Gradients (HOG) features.
# features = hog.compute(img): Computes the HOG features for the input image using the hog object. The features variable will contain the HOG descriptor.
# return features.flatten(): Flattens the HOG feature vector into a one-dimensional array and returns it.
import cv2

def image_to_hog(image_path):
    img = cv2.imread(image_path, 0)  # Load as grayscale
    hog = cv2.HOGDescriptor()
    features = hog.compute(img)
    return features.flatten()

image_path="/content/DS_1.jpeg"
image_vector=image_to_hog(image_path)
print(image_vector)

plt.plot(image_vector)
plt.show()
```

```python
import numpy as np
from PIL import Image

def image_to_matrix(image_path):
  """Converts an image to a NumPy matrix.

  Args:
    image_path: Path to the image file.

  Returns:
    A NumPy array representing the image.
  """

  img = Image.open(image_path)
  img_array = np.array(img)
  return img_array

# Example usage:
image_path = "/content/DS_1.jpeg"
image_matrix = image_to_matrix(image_path)
print(image_matrix.shape)  # Print the dimensions of the matrix
```

```python
# Define the function: video_to_numpy takes a video path as input and returns a NumPy array containing the video frames.
# Create a video capture object: cv2.VideoCapture opens the video file.
# Extract frames: Iterates through the video frames, reading each frame into a NumPy array and appending it to the frames list.
# Convert to NumPy array: Converts the frames list into a NumPy array for further processing.
# Return the video data: Returns the NumPy array containing the video frames.

import cv2
import numpy as np

def video_to_numpy(video_path):
  """Converts a video into a NumPy array of frames.

  Args:
    video_path: Path to the video file.

  Returns:
    A NumPy array of shape (num_frames, height, width, channels).
  """

  cap = cv2.VideoCapture(video_path)
  frames = []

  while True:
    ret, frame = cap.read()
    if not ret:
      break
    frames.append(frame)

  cap.release()
  return np.array(frames)

# Example usage
video_path = "/content/2098988-uhd_3840_2160_30fps.mp4"
video_data = video_to_numpy(video_path)
print(video_data.shape)
# Output: (num_frames, height, width, channels)
```
